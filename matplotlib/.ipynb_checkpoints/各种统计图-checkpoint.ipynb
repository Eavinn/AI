### 3、柱状图

特点：绘制连离散的数据,能够一眼看出各个数据的大小,比较数据之间的差别。(统计/对比)

```python
movie_name = ['雷神3：诸神黄昏','正义联盟','东方快车谋杀案','寻梦环游记','全球风暴','降魔传','追捕','七十七天','密战','狂兽','其它']
x = range(len(movie_name))
y = [73853,57767,22354,15969,14839,8725,8716,8318,7916,6764,52222]
plt.figure(figsize=(20, 8), dpi=100)
plt.bar(x, y, width=0.3, color=['b','r','g','y','c','m','y','k','c','g','b'])
plt.xticks(x, movie_name)
plt.grid(linestyle="--", alpha=0.5)
plt.title("电影票房收入对比")
plt.show()
```

```python
movie_name = ['雷神3：诸神黄昏','正义联盟','寻梦环游记']
first_day = [10587.6,10062.5,1275.7]
first_weekend=[36224.9,34479.6,11830]
x1 = range(len(movie_name))
x2 = [i+0.2 for i in x1]
plt.figure(figsize=(20, 8), dpi=100)
plt.bar(x1, first_day, width=0.2, label="首日票房")
plt.bar(x2, first_weekend, width=0.2, label="首周票房")
plt.legend()
plt.xticks([i+0.1 for i in x1], movie_name)
plt.show()
```

### 5、饼图

特点：分类数据的占比情况(占比)

```python
movie_name = ['雷神3：诸神黄昏','正义联盟','东方快车谋杀案','寻梦环游记','全球风暴','降魔传','追捕','七十七天','密战','狂兽','其它']

place_count = [60605,54546,45819,28243,13270,9945,7679,6799,6101,4621,20105]
plt.figure(figsize=(18,18), dpi=100)
plt.pie(place_count, labels=movie_name,autopct="%1.2f%%", shadow=True,colors=['b','r','g','y','c','m','y','k','c','g','y'])
plt.legend(loc="upper right")
plt.title("电影排片占比")
plt.show()
```
