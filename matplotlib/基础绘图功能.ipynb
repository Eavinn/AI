## matplotlib三层结构：
1. 容器层：
容器层主要由Canvas、Figure、Axes组成。
Canvas（画板）位于最底层；
Figure（画布）建立在Canvas之上，指整个图形(可以通过plt.figure()设置画布的大小和分辨率等)；
Axes（绘图区）建立在Figure之上。
2. 辅助显示层：
Axes(绘图区)内的除了根据数据绘制出的图像以外的内容，主要包括Axes外观(facecolor)、边框线(spines)、坐标轴(axis)、坐标轴名称(axis label)、坐标轴刻度(tick)、坐标轴刻度标签(tick label)、网格线(grid)、图例(legend)、标题(title)等。
3. 图像层：
图像层指Axes内通过plot、scatter、bar、histogram、pie等函数根据数据绘制出的图像。

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
```

### 1. 折线图的绘制和显示

```python
plt.figure(figsize=(20, 8), dpi=100)
plt.plot([1, 2, 3, 4, 5, 6 ,7], [17, 17, 18, 15, 11, 11, 13])
# show之后不能进行图片保存
plt.savefig("test1.png")
plt.show()
```

### 2. 优化辅助显示层

```python
import random
plt.figure(figsize=(15, 4), dpi=100)
x = range(60)
y = [random.uniform(15, 18) for i in range(60)]
plt.plot(x, y)

# 修改刻度，添加中文显示
xticks_label = ["11点{}分".format(i) for i in x[::5]]
plt.xticks(x[::5], xticks_label)
plt.yticks(range(40)[::5])

# 添加网格显示
plt.grid(True, linestyle="--", alpha=0.5)

# 添加描述信息
plt.title("中午11点0分到12点之间的温度变化图示")
plt.xlabel("时间")
plt.ylabel("温度")

plt.show()
```

### 3. 优化图像层

```python
import random

x = range(60)
y_shanghai = [random.uniform(15, 18) for i in range(60)]
y_beijing = [random.uniform(1, 3) for i in x]

plt.figure(figsize=(15, 4), dpi=100)
# 使用多次plot可以画多个折线
plt.plot(x, y_shanghai, label="上海")
plt.plot(x, y_beijing, color='r', linestyle='--', label="北京")
# 显示图例
plt.legend(loc="best")
# 修改刻度，添加中文显示
xticks_label = ["11点{}分".format(i) for i in x[::5]]
plt.xticks(x[::5], xticks_label)
plt.yticks(range(40)[::5])

# 添加网格显示
plt.grid(True, linestyle="--", alpha=0.5)

# 添加描述信息
plt.title("中午11点0分到12点之间的温度变化图示")
plt.xlabel("时间")
plt.ylabel("温度")

plt.show()
```

### 4. 多个坐标系

```python
import random

x = range(60)
y_shanghai = [random.uniform(15, 18) for i in range(60)]
y_beijing = [random.uniform(1, 3) for i in x]

figure, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 4), dpi=100)

axes[0].plot(x, y_shanghai, label="上海")
axes[1].plot(x, y_beijing, color='r', linestyle='--', label="北京")

axes[0].legend(loc="best")
axes[1].legend(loc="best")

xticks_label = ["11点{}分".format(i) for i in x[::5]]
axes[0].set_xticks(x[::5])
axes[0].set_xticklabels(xticks_label)
axes[0].set_yticks(range(40)[::5])
axes[1].set_xticks(x[::5])
axes[1].set_xticklabels(xticks_label)
axes[1].set_yticks(range(40)[::5])

axes[0].grid(True, linestyle="--", alpha=0.5)
axes[1].grid(True, linestyle="--", alpha=0.5)

# 添加描述信息
axes[0].set_title("中午11点0分到12点之间的温度变化图示")
axes[0].set_xlabel("时间")
axes[0].set_ylabel("温度")
axes[1].set_title("中午11点0分到12点之间的温度变化图示")
axes[1].set_xlabel("时间")
axes[1].set_ylabel("温度")

plt.show()
```

### 5. 各种数学图像

```python
import numpy as np
x1 = np.linspace(-10, 10, 10000)
x2 = np.linspace(0.01, 10, 10000)
y1 = np.sin(x1)
y2 = x1 ** 2
y3 = np.log(x2)
y4 = x1 ** 3


_, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 4), dpi=100)
axes[0][0].plot(x1, y1)
axes[0][0].grid()
axes[1][0].plot(x1, y2)
axes[0][1].plot(x2, y3)
axes[0][1].grid()
axes[1][1].plot(x1, y4)

plt.show()
```
