```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import mplfinance as mpf
import talib
```

## 1. 金融数据处理、绘图

```python
stock_day = pd.read_csv("stock_day.csv")
stock_day.sort_index(inplace=True)
stock_day.index = pd.to_datetime(stock_day.index)
stock_day = stock_day[['open', 'close', 'high', 'low', 'volume']]
```

```python
stock_day.head(8)
```

```python
# resample重采样日k-周k，last指最后一天价格
period_week_data = stock_day.resample('w').last()
period_week_data['open'] = stock_day['open'].resample('w').first()
period_week_data['high'] = stock_day['high'].resample('w').max()
period_week_data['low'] = stock_day['low'].resample('w').min()
period_week_data['low'] = stock_day['low'].resample('w').min()
period_week_data['volume'] = stock_day['volume'].resample('w').sum()
period_week_data.dropna(axis=0, inplace=True)
period_week_data
```

```python
# 画日k线
mpf.plot(stock_day[:50])
```

```python
# 画周k线
# 调用make_marketcolors函数，定义K线颜色
mc = mpf.make_marketcolors(
    up="red",  # 上涨K线的颜色
    down="green",  # 下跌K线的颜色
    edge="black",  # 蜡烛图箱体的颜色
    volume="blue",  # 成交量柱子的颜色
    wick="black"  # 蜡烛图影线的颜色
)
style = mpf.make_mpf_style(base_mpl_style="ggplot", marketcolors=mc)
mpf.plot(period_week_data, datetime_format='%Y-%m-%d',type='candle',mav=(5,10,20), 
         volume=True, style=style, figratio=(12, 6))
```

## 2. 时间序列分析

```python
# 移动平均线, 画出股票5,10,20周均线
plt.figure(figsize=(20,8))
period_week_data['close'].rolling(5).mean().plot()
period_week_data['close'].rolling(10).mean().plot()
period_week_data['close'].rolling(20).mean().plot()
plt.show()
```

```python
# 加权移动平均线 (WMA)将过去某特定时间内的价格取其平均值，它的比重以平均线的长度设定，愈近期的收市价，对市况影响愈重要。
# 指数平滑移动平均线(EWMA)
plt.figure(figsize=(20,8))
pd.DataFrame.ewm(period_week_data['close'],span=5).mean().plot()
pd.DataFrame.ewm(period_week_data['close'],span=10).mean().plot()
pd.DataFrame.ewm(period_week_data['close'],span=20).mean().plot()
plt.show()
```

```python
# 移动方差var和标准差std
plt.figure(figsize=(20,8))
# period_week_data['close'].rolling(5).std().plot()
period_week_data['close'].rolling(5).var().plot()
period_week_data['close'].rolling(10).var().plot()
period_week_data['close'].rolling(20).var().plot()
plt.show()
```

## 3. MACD

快线：12日移动平均值-26日移动平均值。
DIF(离差值)=EMA(close,12)-EMA(close,26)

慢线：9日DIF移动平均值。
DEA(讯号线)=EMA(DIF,9)

柱状图：BAR=2*(DIF-DEA)

```python
period_week_data
```

```python
# 计算快线、慢线、量柱
dif, dea, macd_hist = talib.MACD(period_week_data['close'].values)
red_hist = np.where(macd_hist > 0, macd_hist, 0)
green_hist = np.where(macd_hist < 0, macd_hist, 0)
```

```python
# 添加macd子图
add_plot = [mpf.make_addplot(dif, panel=2, color='r'),
            mpf.make_addplot(dea, panel=2, color='g'),
            mpf.make_addplot(red_hist, type='bar', width=0.7, panel=2, color='r'),
            mpf.make_addplot(green_hist, type='bar', width=0.7, panel=2, color='g')]
            
mpf.plot(period_week_data, type='candle', datetime_format='%Y-%m-%d', figratio=(12, 6), style=style,
         volume=True, addplot=add_plot)
```

```python
# 实现MACD交易策略
for i in range(1, period_week_data.shape[0]):
    if dif[i] > dea[i] and dif[i-1] <= dea[i-1]:
        print("日期%s应买入股票" % period_week_data.index[i].strftime("%Y-%m-%d"))
    if dif[i] < dea[i] and dif[i-1] >= dea[i-1]:
        print("日期%s应卖出股票" % period_week_data.index[i].strftime("%Y-%m-%d"))
```

## 3. RSI

RSI=100×RS/(1+RS)

RS=14天内收市价上涨数之和的平均值/14天内收市价下跌数之和的平均值

```python
# 计算RSI长短线
short_rsi = talib.RSI(period_week_data['close'].values, timeperiod=6)
long_rsi = talib.RSI(period_week_data['close'].values, timeperiod=12)
```

```python
# 实现RSI交易策略
for i in range(1, period_week_data.shape[0]):
    if short_rsi[i] < 20 and short_rsi[i] > long_rsi[i] and short_rsi[i-1] <= long_rsi[i-1]:
        print("日期%s应买入股票" % period_week_data.index[i].strftime("%Y-%m-%d"))
    if short_rsi[i] > 80 and short_rsi[i] < long_rsi[i] and short_rsi[i-1] >= long_rsi[i-1]:
        print("日期%s应卖出股票" % period_week_data.index[i].strftime("%Y-%m-%d"))
```

## 4. BOLL

中轨线(MA)=N日的移动平均线

MB=（N－1）日的MA

上轨线(UP)=中轨线＋两倍的标准差

下轨线(DN)=中轨线－两倍的标准差

```python
upperband, middleband, lowerband = talib.BBANDS(period_week_data['close'].values)
```

```python
# 添加BOLL线
add_plot = [mpf.make_addplot(upperband, color='g'),
            mpf.make_addplot(middleband, color='k'),
            mpf.make_addplot(lowerband, color='b')]
            
mpf.plot(period_week_data, type='candle', datetime_format='%Y-%m-%d', figratio=(12, 6), style=style,
         volume=True, addplot=add_plot)
```
